```python
# Imports required for this exercise
from dotenv import load_dotenv
import os
from openai import OpenAI
from IPython.display import Markdown,display
```

```python
# load the env variables from ".env" file
load_dotenv(override=True)
```

```python
# Check if the API key is set
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("OPENAI_API_KEY is not set!!")
```

```python
# Create an instance of the openai python client
openai_client = OpenAI()
```

```python
# First ask LLM to identify the most important pain point in the Cybersecurity domain that companies need to
# focus on in 2026
message = """
What is one top area for cybersecurity organizations to focus on in 2026? Limit your answer to 3-5 sentences.
Only return the top pain point and no additional details on solution. Respond in markdown without code blocks.
"""
messages = [
    {"role": "user", "content": message}
]

MODEL_NAME="gpt-4o-mini"

response = openai_client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages
)

painpoint = response.choices[0].message.content
display(Markdown(painpoint))
```
