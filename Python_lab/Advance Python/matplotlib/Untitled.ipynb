```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,5,7,8]
y=[10,20,30,20,35,40,20,10,50]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color="red")
```

```python
plt.plot(x,y, color="blue",linestyle="dotted",linewidth=5)
```

```python
plt.xlabel("Day")
plt.ylabel("Range")
plt.title("Temperature")
plt.plot(x,y, "--*k",linewidth=10)
```
