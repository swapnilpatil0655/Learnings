```python
# bar chart
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ["Google", "Amazi", "MsFT", "Wipro"]  
revenue = [90, 136, 89, 27]
```

```python
plt.bar(company, revenue) 
```

```python
ypos = np.arange(len(company))
ypos
```

```python
#plt.xticks(ypos, company) #xticks having 2arguments
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company) #xticks having 2arguments
plt.ylabel("revenue")
plt.xlabel("company_name")
plt.title("it companys")
plt.bar(ypos, revenue, label="revenue")
plt.legend()
plt.show()
```

```python
#now we also adding profit 
company = ["Google", "Amazi", "MsFT", "Wipro"]
revenue = [90, 136, 89, 27]
profit = [40, 15, 34, 12]
```

```python
plt.xticks(ypos, company) #xticks having 2arguments
plt.ylabel("revenue")
plt.xlabel("company_name")
plt.title("it companys")
plt.bar(ypos, revenue, label="revenue")
plt.bar(ypos, profit, label = "Profit")
plt.legend()
# why both are at same chat , because bothhaving x position issame
```

```python
plt.xticks(ypos, company) #xticks having 2arguments
plt.ylabel("revenue")
plt.xlabel("company_name")
plt.title("it companys")
plt.bar(ypos, revenue, label="revenue")
plt.bar(ypos, profit, label = "Profit")
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos, company) #xticks having 2arguments
plt.ylabel("revenue")
plt.xlabel("company_name")
plt.title("it companys")
plt.bar(xpos-0.2, revenue,width=0.4, label="revenue")
plt.bar(xpos+0.2, profit,width=0.4, label = "Profit")
plt.legend()
```

```python
# horizontal bar chart
plt.yticks(ypos, company) #xticks having 2arguments
#plt.ylabel("revenue")
#plt.xlabel("company_name")
plt.title("it companys")
plt.barh(xpos-0.2, revenue, label="revenue")
plt.barh(xpos+0.2, profit, label = "Profit")
plt.legend()
```

```python
blood_sugar_man = [113, 85, 150, 149, 88,93,115,135,80,77,82,129]
blood_sugar_woman=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel("sugar range")
plt.ylabel("no. of patients")
plt.title("Blood sugar analysis")
plt.hist([blood_sugar_man,blood_sugar_woman], bins=[75, 100,125,150], rwidth=0.80, color =["r", "b"],
        label=["man", "woman"])
plt.legend()
```

```python
#pai chart 

exp_value=[1400,600,300,410,250]
exp_cate = ["home rant", "Food","phone bill", "car","other expensis"]
```

```python
plt.pie(exp_value, labels=exp_cate)
```

```python
plt.axis("equal")
plt.pie(exp_value, labels=exp_cate, radius=1)
plt.show()
```

```python
# percentages
plt.axis("equal")
plt.pie(exp_value, labels=exp_cate, radius=1, autopct="%0.3f%%", shadow=True)
plt.show()
```

```python
# percentages
plt.axis("equal")
plt.pie(exp_value, labels=exp_cate, radius=1, autopct="%0.1f%%", explode=[0.1,0.3,0.2,0.2,0.2]
        )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_value, labels=exp_cate, radius=1, autopct="%0.1f%%", explode=[0,0,0,0,0], 
        startangle=180)
plt.show()
plt.savefig("piechart.png", bbox_inches ="tight")
```
