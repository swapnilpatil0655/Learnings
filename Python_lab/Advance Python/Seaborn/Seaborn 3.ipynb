```python
plt.figure(figsize = (14,12))
sns.set(style="darkgrid")
sns.lineplot(x="size", y="tip", data=tips_df, hue="sex", style= "sex", palette="hot", dashes=True,
            markers=["o", "<"], legend="brief",)
plt.title("line_plot", fontsize=20)
plt.xlabel("size", fontsize= 15)
plt.ylabel("Total Bill", fontsize=15)
```

```python
tips_df.value_counts("day")
```

```python
plt.figure(figsize = (14,12))
sns.set(style="darkgrid")
sns.lineplot(x="size", y="tip", data=tips_df, hue="day", style= "day", palette="hot", dashes=True,
            markers=["o", "<", ">", "^"], legend="brief",)
plt.title("line_plot", fontsize=20)
plt.xlabel("size", fontsize= 15)
plt.ylabel("Total Bill", fontsize=15)
```

# How to draw Histogram in Seaborn/ seaborn Distplot


```python
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df.value_counts("day")
```

```python
tips_df
```

```python
sns.distplot(tips_df.value_counts("day"))
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"], hist= True)
```

```python
sns.distplot(tips_df["total_bill"], kde=False)
```

```python
from scipy.stats import norm
import seaborn as sns
sns.distplot(tips_df["total_bill"], fit = norm)
```

```python

sns.distplot(tips_df["total_bill"], color = "r")
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(tips_df["total_bill"], bins=bins)
plt.xticks(bins)
plt.title("histofram of total bill")
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()

sns.distplot(tips_df["total_bill"], hist_kws= {"color":"#DC143C", "edgecolor": "#aaff00", "linewidth":5,
                                              "linestyle":"--", "alpha":0.9})
plt.xticks(bins)
plt.title("histofram of total bill")
plt.show()
```
