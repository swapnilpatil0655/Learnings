# Corellation with heatmap

 Correlation heatmaps are a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

# join plot
a join plot allows to study the relationship between 2 numeric variables

```python
# for using univariate analysis
sns.jointplot(x="tip", y="total_bill", data= tips, kind="hex")
```

```python
sns.jointplot(x="tip", y="total_bill", data= tips, kind="reg")
```

# Pair plot 
A psir plot is also know as a scatterplot, in which, one variable in the same data row is matched with another variables
value, like this, pair plots are just elaborations on this showing all variables paired with all the other variables.


```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex")
```

```python
#tips["smoker"].value_count()
sns.pairplot(tips, hue="smoker")
```

# Dist plot
Disht plot helps us to check the distribution of the columns feature

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["tip"],kde=False, bins=10) # kde kernal density estimation
```

# Categorical plot
seaborn alsohelps us in doing the analysis on Cateforical Data points, in this section we will discuss.

boxlpot
violinplot
countplot
bar plot


```python
sns.countplot("sex", data=tips) # bydefault it taking x axis
```

```python
sns.countplot("day", data=tips)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
# bar plot
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
# Box plot
sns.boxplot("sex", "total_bill", data=tips)
```

```python
sns.boxplot(x="day",y= "total_bill", data=tips, palette="rainbow")
```

```python
sns.boxplot(data=tips, orient="v")
```

```python
# categorize my data based on some other categories
sns.boxplot(x="total_bill", y="day", hue="smoker", data=tips)
```

```python
# Violine plot
sns.violinplot(x="total_bill", y="day", data=tips, palette= "rainbow")
```

```python
import seaborn as sns
sns.set_theme(style="white")

# Load the example mpg dataset
mpg = sns.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=mpg)
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

df = sns.load_dataset("penguins")
sns.pairplot(df, hue="species")
```

```python
#Scatterplot with categorical variables
import seaborn as sns
sns.set_theme(style="whitegrid", palette="muted")

# Load the penguins dataset
df = sns.load_dataset("penguins")

# Draw a categorical scatterplot to show each observation
ax = sns.swarmplot(data=df, x="body_mass_g", y="sex", hue="species")
ax.set(ylabel="")
```
