```python
pin=1234
Balance=10000
flag=True
count=0

while flag:
    upin=int(input("Enter the ATM Pin"))
    if (upin==pin):
        print("you have succesfully login")
        print("Press 1 for Balance check")
        print("Press 2 for Withdrown")
        print("Press 3 for Deposite")
        print("Press 4 for Pin change")
        print("press 5 for Exit")
        #enter the output #
        opt=int(input())
        if(opt==1):
            print("Balance:",Balance)
        elif(opt==2):
            Withdrown=int(input("Withdrown"))
            Balance=Balance-Withdrown
            print("Withdrown a",Withdrown)
            print("Balance",Balance)
        elif(opt==3):
            Deposite=int(input("Deposite"))
            Balance=Balance+Deposite
            print("Deposite a",Deposite)
            print("Balance",Balance)
        elif(opt==4):
            new_pin=int(input("Enter the new pin"))
            if(new_pin==pin):
                print("your new pin is similar as old pin , please try again")
                new_pin=int(input())
            else:
                pin=new_pin
                print("your pin is succesfully change")
        elif(opt==5):
            flag=False
    else:
        print("you have entered wrong pin please enter again")
        count=count+1
        if(count==3):
            print("your card is blocked ")
            flag=False
```
