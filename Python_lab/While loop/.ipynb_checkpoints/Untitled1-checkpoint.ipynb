```python
class exit():
    def getexit(self):
        print("Thank you for visiting ")

class balance():
    bal=1000
    def update(self,amt,opt):
        if opt==1:
            print("before deposite amount is ",self.bal)
            self.bal=self.bal+amt
            print("after deposite amount is ",self.bal)
            
            e=exit()
            e.getexit()
            
        elif opt==2:
            print("before withdraw amount is ",self.bal)
            self.bal=self.bal-amt
            print("after withdraw amount is ",self.bal)
        elif opt==3:
            print("Your total balance is ", self.bal)
class withdraw():
    def getwithdraw(self,amt,opt):
        b=balance()
        b.update(amt,opt)
        

class deposite():
    def getdeposite(self,amt,opt):
        b=balance()
        b.update(amt,opt)

class menu():
    def getmenu(self):
        print("press 1 for deposite ")
        print("press 2 for withdraw ")
        print("print 3 for balance ")
        opt=int(input("Enter the option "))
        if opt==1:
            d=deposite()
            amt=int(input("Enter the amount : "))
            d.getdeposite(amt,opt)
        elif opt==2:
            w=withdraw()
            amt=int(input("Enter the amount"))
            w.getwithdraw(amt,opt)
        elif opt==3:
            b=balance()
            amt=0
            b.update(amt,opt)
class authentication():
    pin=1234
    def auth(self,upin):
        
        if upin==self.pin:
            print("Login successfully ")
            m=menu()
            m.getmenu()
        else:
            print("Login failed ")
        
class bank():
    def welcome(self):
        print("Welcome to ATM ")
        upin=int(input("enter 4 digit pin : "))
        a=authentication()
        a.auth(upin)
        
b=bank()
b.welcome()
```

```python
class bank():
    def welcome(self):
        print("welcome to the ATM")
        upin=int(input("enter the pin"))
        a=auth()
        a.getauth(upin)
        
class auth():
    def getauth(self,upin):
        pin=1234
        if pin==upin:
            print("login successfull")
            m=menu()
            m.getmenu()
        else:
            print("login failed")

class menu():
    def getmenu(self):
        print("press 1 for deposite ")
        print("press 2 for withdraw ")
        print("press 3 for balance check ")
        opt=int(input("Enter the option "))
        if opt==1:
            d=deposite()
            d.getdeposite(opt)
        elif opt==2:
            w=wit()
            w.getwit(opt)
        elif opt==3:
            e=enq()
            e.getenq(opt)
            
class deposite():
    def getdeposite(self,opt):
        amt=int(input("Enter the amount : "))
        b=balance()
        b.getbalance(amt,opt)

class wit():
    def getwit(self,opt):
        amt=int(input("Enter the amount : "))
        b=balance()
        b.getbalance(amt,opt)
        
class enq():
    def getenq(self,opt):
        amt=0
        b=balance()
        b.getbalance(amt,opt)
        
class balance():
    def getbalance(self,amt,opt):
        balance=10000
        if opt==1:
            print("before balance : ",balance)
            balance=balance+amt
            print("after balance : ",balance)
        elif opt==2:
            print("before balance : ",balance)
            balance=balance-amt
            print("after balance : ",balance)
        elif opt==3:
            print("before balance : ",balance)
            balance=balance+amt
            print("after balance : ",balance)
        
b=bank()
b.welcome()
```
