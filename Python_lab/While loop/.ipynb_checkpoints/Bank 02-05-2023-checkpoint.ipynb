```python
class balance():
    bal=10000
    def update(self,amt,opt):
        if(opt==1):
            print("balance before deposite",self.bal)
            self.bal=self.bal+amt
            print("balance after deposite",self.bal)
        elif(opt==2):
            print("balance before withdraw",self.bal)
            self.bal=self.bal-amt
            print("balance after withdraw ",self.bal)
    def getcurrentbalance(self):
        return self.bal
    
class withdraw():
    def setwithdraw(self,amt,opt,obj):
        obj.getbalance().update(amt,opt)
        


class deposite():
    def setdeposite(self,amt,opt,obj):
        obj.getbalance().update(amt,opt)
        


class menu():
    def display(self,obj):
        flag=True
        while flag:
            print("press 1 for deposite ")
            print("press 2 for withdraw")
            print("press 3 for balance ")
            print("press 4 for exit ")
            opt=int(input("Enter your option : "))
            if opt==1:
                amt=int(input("Enter amount to deposite : "))
                obj.getdeposite().setdeposite(amt,opt,obj)
            elif opt==2:
                amt=int(input("Enter amount for withdraw"))
                obj.getwithdraw().setwithdraw(amt,opt,obj)
            elif opt==3:
                print(obj.getbalance().getcurrentbalance())

            elif opt==4:
                print("Thank you for visiting.....")
                flag==False
                break
                
            
    
        
        

class database():
    def getpin(self):
        return 1234

class authenticate():
    def auth(self,upin,obj):
        if (upin==obj.getdatabase().getpin()):
            print("login successful")
            obj.getmenu().display(obj)

class bank():
    def welcome(self,obj):
        print("Welcome to the ATM ")
        upin=int(input("Enter your pin : "))
        obj.getauthenticate().auth(upin,obj)
        
class ObjectFactory():
    b=bank()
    a=authenticate()
    d=database()
    m=menu()
    dep=deposite()
    bal=balance()
    wit=withdraw()
    def getwithdraw(self):
        return self.wit
    def getbalance(self):
        return self.bal
    def getdeposite(self):
        return self.dep
    def getmenu(self):
        return self.m
    def getdatabase(self):
        return self.d
    def getauthenticate(self):
        return self.a
    def getbank(self):
        return self.b
    
    
obj=ObjectFactory()
x=obj.getbank()
x.welcome(obj)
```
