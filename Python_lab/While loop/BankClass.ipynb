```python
class Balance():
    bal=10000
    def update(self, amt, opt):
        if(opt==1):
            print("bal before deposite", self.bal)
            self.bal=self.bal+amt
            print("bal after deposite", self.bal)
        elif(opt==2):
            print("bal before withdraw", self.bal)
            self.bal=self.bal-amt
            print("bal after withdraw", self.bal)
    def getCurrentBalance(self):
        return self.bal
class Deposite():
    def getDeposite(self, amt, opt, obj):
        obj.getBalance().update(amt, opt)
class Withdraw():
    def getWithdraw(self, amt, opt, obj):
        obj.getBalance().update(amt, opt)
    
        
class Menu():
    def display(self,obj):
        flag=True
        while(flag):
            print("Press 1 for Deposirte")
            print("press 2 for withdraw")
            print("press 3 for Balance")
            print("Press 4 for Exit")
            opt=int(input("Enter ur option"))
            if(opt==1):
                amt=int(input("Enter amt to deposite"))
                obj.getDepositeObject().getDeposite(amt,opt,obj)
            elif(opt==2):
                amt=int(input("Enter amt to Withdraw"))
                obj.getWithdrawObject().getWithdraw(amt,opt,obj)
            elif(opt==3):
                print(obj.getBalance().getCurrentBalance())
            elif(opt==4):
                flag=False
                print("Thank u for banking with us")
            else:
                print("wrong option")
            
            
class DataBase():
    def getPin(self):
        return 1234
    
class Authenticate():
    def auth(self,upin,obj):  
       
        if(upin==obj.getDataBase().getPin()):
            print("Login Succesful")
            obj.getMenu().display(obj)
        else:
            print("login failure")

class Bank():
    def welcome(self,obj):
        print("Welcome to the Bank")
        upin=int(input("enter ur 4 digit pin"))
        obj.getAuthenticate().auth(upin,obj)

class ObjectFactory():
    b=Bank()
    d=DataBase()
    a=Authenticate()
    m=Menu()    
    dep=Deposite()
    bal=Balance()
    wit=Withdraw()
    def getWithdrawObject(self):
        return self.wit
    def getBalance(self):
        return self.bal
    def getDepositeObject(self):
        return self.dep
    def getBank(self):
        return self.b
    def getDataBase(self):
        return self.d
    def getAuthenticate(self):
        return self.a
    def getMenu(self):
        return self.m
    
obj=ObjectFactory()
x=obj.getBank()
x.welcome(obj)
#obj.getBank().welcome()
```
