```python
## Assignment
## Simple Calculator program
# Take user input

num_1 = float(int(input("Enter first number: ")))
num_2 = float(int(input("Enter second number: ")))
operation = input("Enter operation (+.-,*,/): ")
# Perform calculation based on operation
if operation == "+":
    result = num_1 + num_2
elif operation =="-":
    result = num_1 - num_2
elif operation == "*":
    result = num_1 * num_2
elif operation == "/":
    if num_1 or num_2 == 0:
        print("Error! Division by zero.")
    else:
        result = num_1 / num_2

else:
    result = "Invalid operation"

print(result)
```
